# inserting the data set and and seeing whether it contains any missing values, duplicates,outliers

```python
sales_data = pd.read_csv("/content/May-2022.csv")
```

```python
# Count missing values using isna()
missing_values_count = sales_data.isna().sum()
print("\nMissing values in each column using isna():\n", missing_values_count)
```

```python
# Visualize missing values
sns.heatmap(sales_data.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

* **this data has no missing values **

# Now we have go for the duplicates

```python
# Check for duplicates
duplicates = sales_data.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

# Remove duplicate rows
sales_data = sales_data.drop_duplicates()
```

**The data set has no duplicate values**
